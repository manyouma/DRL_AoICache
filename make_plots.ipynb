```python
import scipy.io as sio
import numpy as np
from matplotlib import pyplot as pl

DELTA_VALUE = 5
N_USERS = 2
ETA_value = 5
convergence_vec_5 = np.zeros((10,200))
convergence_vec_15 = np.zeros((10,200))
convergence_vec_25 = np.zeros((10,200))
for RANDOM_SEED in range(10):
    file_name = 'data/AoI_Eta%d_Random%d_Delta%d_N_User%d.mat'%(5,RANDOM_SEED,DELTA_VALUE,N_USERS)
    A_temp = sio.loadmat(file_name)
    convergence_vec_5[RANDOM_SEED,:] = A_temp['convergence_vec']
    file_name = 'data/AoI_Eta%d_Random%d_Delta%d_N_User%d.mat'%(15,RANDOM_SEED,DELTA_VALUE,N_USERS)
    A_temp = sio.loadmat(file_name)
    convergence_vec_15[RANDOM_SEED,:] = A_temp['convergence_vec']
    file_name = 'data/AoI_Eta%d_Random%d_Delta%d_N_User%d.mat'%(25,RANDOM_SEED,DELTA_VALUE,N_USERS)
    A_temp = sio.loadmat(file_name)
    convergence_vec_25[RANDOM_SEED,:] = A_temp['convergence_vec']
```

```python
x = np.linspace(0, 30, 30)
y = np.sin(x/6*np.pi)
error = np.random.normal(0.1, 0.02, size=y.shape) +.1
y += np.random.normal(0, 0.1, size=y.shape)

pl.plot(x, y, 'k', color='#CC4F1B')
pl.fill_between(x, y-error, y+error,
    alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848')
print(x.size)
print(y.size)

pl.show()
```

```python
x = np.linspace(0,199,200)
y_5 = np.mean(-convergence_vec_5,0)
error_5 = np.var(-convergence_vec_5,0)
pl.plot(x, y_5, 'k', color='#CC4F1B')
pl.fill_between(x, y-error_5, y+error_5,
    alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848')


y_15 = np.mean(-convergence_vec_15,0)
error_15 = np.var(-convergence_vec_15,0)
pl.plot(x, y_15, 'k', color='#1B2ACC')
pl.fill_between(x, y-error_15, y+error_15,
    alpha=0.2, edgecolor='#1B2ACC', facecolor='#089FFF')

y_25 = np.mean(-convergence_vec_25,0)
error_25 = np.var(-convergence_vec_25,0)
pl.plot(x, y_25, 'k', color='#3F7F4C')
pl.fill_between(x, y-error_25, y+error_25,
    alpha=1, edgecolor='#3F7F4C', facecolor='#7EFF99')


pl.ylim((0,50))
pl.xlim((0,50))
pl.show()
```
